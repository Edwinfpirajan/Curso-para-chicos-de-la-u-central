Matrices Inversas
Si una matriz cuadrada  $A$  de $ n×n$   se multiplica por derecha e izquierda por una matriz  $B$  y se obtiene la identidad en ambos casos, es decir
                         $ AB=BA=In $  
La matriz $ B $ se conoce como el inverso multiplicativo de la matriz $ A $
 o más comunmente como la matriz inversa de  $A$ .

No todas las matrices cuadradas tienen inversa, la forma de determinarla empleando Python es meduante la función np.linalg.inv



```python
D= np.array ([[1,2],[3,4]])
inv=np.linalg.inv (D)
inv
```

```python
D= np.array ([[1,2,3],[0,4,5],[0,1,2]])
inv=np.linalg.inv (D)
inv
```

```python
D@inv #comprobacion
```

```python
D =np.array([[1,2],[3,4]])
inversa=np.array([[-2,1],[1.5,-0.5]])
```

```python
D@inversa
```

# Determinante de una matriz 
el determinate de una matriz $A$ es un numero real asociado a la matriz que perimite determinar si una matriz es invertible o no lo es.

Una matriz $A$ es invertible si solamnte si su determinante es distinto de cero.

En otras palabras una matriz $A$ es **singular** si su determinate es cero.

Utilizamos el comando **np.linag.det(nombre de matriz) ** para calcular el determiante

```python
c=np.array([[1,2],[3,4]])
np.linalg.det (c)
```

```python
C1 =np.array([[10,8],[5,4]])
np.linalg.det (C1)
```
